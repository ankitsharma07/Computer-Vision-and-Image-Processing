# Applying filter to an image

I will use `scikit-image` library for this experiment

```python
from skimage import io, color
%config InlineBackend.figure_format = 'retina'
```

```python
#Check the dimension of the image
img = io.imread('Pictures/pug.png')
dimensions = color.guess_spatial_dimensions(img)
print(dimensions)
```

## Color to Gray Scale

```python
from skimage.color import rgb2gray
img = io.imread('Pictures/pug.png')
img_grayscale = rgb2gray(img)
```

## Save it as grayscale

```python
io.imsave('pug-gs.png', img_grayscale)
```

## To show the image


```python
show_grayscale = io.imshow(img_grayscale)
io.show()
```

## Applying a filter to this image

```python
from skimage import data, io, filters
```

```python
img = io.imread('Pictures/pug.png')
```

```python
edges = filters.sobel(img) #We are giving a colored image that's why the problem exists.
```

#### Providing the grayscale image 

```python

img = io.imread('Pictures/pug-gs.png')
```

```python
edges = filters.sobel(img)
io.imshow(edges)
io.show()
```
